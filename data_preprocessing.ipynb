```python
import pandas as pd
import re
```

```python
# Tạo danh sách để lưu trữ các giá trị label và name
labels = []
names = []
path = '/Dataset/product_train_0.txt'

# Đọc từng dòng từ tệp văn bản
with open(path, 'r', encoding='utf-8') as file:
    for line in file:
        # Xóa bỏ ký tự xuống dòng ở cuối mỗi dòng
        line = line.strip().lower()

        # Áp dụng quy tắc tách label và name
        if line.startswith('__label__'):
            label, name = line[9:].split(' ', 1)
            labels.append(label)
            names.append(name)

# Tạo DataFrame từ danh sách labels và names
df = pd.DataFrame({
    'Label': labels,
    'Name': names
})
```

```python
df.head()
```

```python
# Hàm làm sạch văn bản và lọc từ không thuộc tiếng Việt
def clean_text(text):
    text = text.lower()  # Chuyển đổi thành chữ thường
    text = re.sub(r'\S*\d+\S*', '', text)  # Loại bỏ các chuỗi con chứa số
    text = re.sub(r'\W+', ' ', text)  # Thay thế các ký tự không phải là chữ cái hoặc số bằng khoảng trắng
    words = text.split()  # Tách chuỗi thành các từ

    return ' '.join(words)  # Kết hợp lại thành chuỗi

df['Name'] = df['Name'].apply(clean_text)
```
