```python
import pandas as pd
import string
import re
import nltk
```

```python
# Tạo danh sách để lưu trữ các giá trị label và name
labels = []
names = []
path = '/content/drive/MyDrive/Colab Notebooks/Test Dagoras/product_train_0.txt'

# Đọc từng dòng từ tệp văn bản
with open(path, 'r', encoding='utf-8') as file:
    for line in file:
        # Xóa bỏ ký tự xuống dòng ở cuối mỗi dòng
        line = line.strip().lower()

        # Áp dụng quy tắc tách label và name
        if line.startswith('__label__'):
            label, name = line[9:].split(' ', 1)
            labels.append(label)
            names.append(name)

# Tạo DataFrame từ danh sách labels và names
df = pd.DataFrame({
    'Label': labels,
    'Name': names
})
```

```python
df
```

```python
def label_clean_text(text):
    # Chuyển đổi thành chữ thường
    text = text.lower()

    # Thay thế '_' bằng ' '
    text = text.replace('_', ' ')

    # Loại bỏ dấu câu và ký tự đặc biệt
    text = re.sub(f"[{string.punctuation}]", "", text)

    return text

# Áp dụng hàm label_clean_text vào cột 'Label' của DataFrame df
df['Label'] = df['Label'].apply(label_clean_text)

# Hàm để chuẩn hóa chuỗi
def normalize_label(label):
    # Loại bỏ khoảng trắng thừa và chuẩn hóa dấu cách
    return ' '.join(label.strip().split())

# Áp dụng hàm normalize_label cho từng phần tử trong cột 'Label'
df['Label'] = df['Label'].apply(normalize_label)
```

```python
df = df.reset_index(drop=True)
df
```
